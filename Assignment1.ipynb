# 19) Min and max monthly charges for customer that has partner group by gender

```python
#function min and max
def get_stats(group):
    return {'min': group.min(), 'max': group.max()}

#group by gender
custPartner = telco[telco['Partner'].isin(['Yes'])]  #select by partner
custPartner['MonthlyCharges'].groupby(custPartner['gender']).apply(get_stats).unstack()
```
